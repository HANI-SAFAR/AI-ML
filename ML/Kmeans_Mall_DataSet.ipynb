**Question 1**:Given a list of tuples, Write a python code to extract all tuples which have all elements divisible by K. Print the explanation

Input : test_list = [(6, 24, 12), (60, 12, 6), (12, 18, 21)], K = 6 
Output : [(6, 24, 12), (60, 12, 6)] 
Explanation : Both tuples have all elements multiple of 6.

```
# initializing list 
test_list=eval(input("Enter the data of dictionaries :"))
K=int(input("Enter the number which is divisible :"))

res = [sub for sub in test_list if all(ele % K == 0 for ele in sub)] 
  
# printing result 
print("K Multiple elements tuples : " + str(res))
```

```
#1-Creating Dataframe
import pandas as pd
import numpy as np
df=pd.read_csv("/content/Mall_Customers.csv")
df
```

```
df.columns
```

```
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaled_df=scaler.fit_transform(df[['Annual Income (k$)', 'Spending Score (1-100)']])
scaled_df
```

```
#2 Implement k means clustering with k=3
from sklearn.cluster import KMeans 
k=3 
cluster=KMeans(k,random_state=42) 
cluster.fit(scaled_df) 
df["clusterid"]=cluster.labels_

df[df["clusterid"]==0]
```

```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

cluster_range = range(1,10) 
cluster_errors = []
for num_clusters in cluster_range: 
  clusters = KMeans(num_clusters) 
  clusters.fit(scaled_df) 
  cluster_errors.append(clusters.inertia_)
plt.figure(figsize = (6,4))
plt.plot(cluster_range, cluster_errors, marker = "*") 
plt.xlabel("No. of clusters")
plt.ylabel("Sum of Squared Error")
```

```
#3 MEAN AND STANDARD DEVIATION FOR EACH CLUSTER
cluster=KMeans(3)
clusters.fit(scaled_df)
df["new_clusterid"]=clusters.labels_
df.groupby("new_clusterid")["Annual Income (k$)","Spending Score (1-100)"].agg(["mean","std"]).reset_index()
```
